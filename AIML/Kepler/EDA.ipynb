```python
import pandas as pd

# Input file
INPUT_FILE = "kepler.csv"

# Load
df = pd.read_csv(INPUT_FILE)

# --- Core feature set ---
core_features = [
    # Labels
    "koi_disposition",

    # Transit geometry
    "koi_period", "koi_period_err1", "koi_period_err2",
    "koi_duration", "koi_duration_err1", "koi_duration_err2",
    "koi_depth", "koi_depth_err1", "koi_depth_err2",
    "koi_ingress", "koi_ingress_err1", "koi_ingress_err2",
    "koi_impact", "koi_impact_err1", "koi_impact_err2",
    "koi_ror", "koi_ror_err1", "koi_ror_err2",
    "koi_srho", "koi_srho_err1", "koi_srho_err2",

    # Planet props
    "koi_prad", "koi_prad_err1", "koi_prad_err2",
    "koi_sma", "koi_sma_err1", "koi_sma_err2",
    "koi_incl", "koi_incl_err1", "koi_incl_err2",
    "koi_teq", "koi_teq_err1", "koi_teq_err2",
    "koi_insol", "koi_insol_err1", "koi_insol_err2",
    "koi_dor", "koi_dor_err1", "koi_dor_err2",

    # Stellar props
    "koi_steff", "koi_steff_err1", "koi_steff_err2",
    "koi_slogg", "koi_slogg_err1", "koi_slogg_err2",
    "koi_smet", "koi_smet_err1", "koi_smet_err2",
    "koi_srad", "koi_srad_err1", "koi_srad_err2",
    "koi_smass", "koi_smass_err1", "koi_smass_err2",

    # Vetting / SNR
    "koi_model_snr", "koi_num_transits",
    "koi_model_chisq", "koi_model_dof",
    "koi_fpflag_nt", "koi_fpflag_ss", "koi_fpflag_co", "koi_fpflag_ec"
]

# --- Optional feature set ---
optional_features = [
    # Scores / dispositions
    "koi_score",

    # Photometric magnitudes
    "koi_kepmag", "koi_gmag", "koi_rmag", "koi_imag", "koi_zmag",
    "koi_jmag", "koi_hmag", "koi_kmag",

    # Centroid / imaging checks
    "koi_fwm_stat_sig", "koi_fwm_sra", "koi_fwm_sra_err",
    "koi_fwm_sdec", "koi_fwm_sdec_err", "koi_fwm_srao", "koi_fwm_srao_err",
    "koi_fwm_sdeco", "koi_fwm_sdeco_err", "koi_fwm_prao", "koi_fwm_prao_err",
    "koi_fwm_pdeco", "koi_fwm_pdeco_err",
    "koi_dicco_mra", "koi_dicco_mra_err", "koi_dicco_mdec", "koi_dicco_mdec_err",
    "koi_dicco_msky", "koi_dicco_msky_err", "koi_dikco_mra", "koi_dikco_mra_err",
    "koi_dikco_mdec", "koi_dikco_mdec_err", "koi_dikco_msky", "koi_dikco_msky_err"
]

# Always keep IDs
id_columns = ["rowid", "kepid", "kepoi_name"]

# Build datasets
core_df = df[id_columns + [col for col in core_features if col in df.columns]]
core_opt_df = df[id_columns + [col for col in (core_features + optional_features) if col in df.columns]]

# Save
core_df.to_csv("kepler_core.csv", index=False)
core_opt_df.to_csv("kepler_core_optional.csv", index=False)

print("✅ Saved kepler_core.csv and kepler_core_optional.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder

# Load
df = pd.read_csv("kepler_core.csv")

# --- 1. Basic info ---
print("Shape:", df.shape)
# print("\nColumns:", df.columns.tolist())
# print("\nHead:\n", df.head())

# --- 2. Target distribution ---
print("\nTarget column distribution (koi_disposition):")
print(df["koi_disposition"].value_counts())

sns.countplot(x="koi_disposition", data=df)
plt.title("Class Balance")
plt.show()

# --- 3. Missing values ---
print("\nMissing values per column:")
print(df.isnull().sum().sort_values(ascending=False))

# --- 4. Encode target for correlation ---
le = LabelEncoder()
df["target_encoded"] = le.fit_transform(df["koi_disposition"])
print("\nLabel mapping:", dict(zip(le.classes_, le.transform(le.classes_))))

# --- 5. Numerical features + target ---
num_cols = df.select_dtypes(include=["float64", "int64"]).columns.tolist()
# Make sure target_encoded is included
if "target_encoded" not in num_cols:
    num_cols.append("target_encoded")

# --- 6. Correlation matrix ---
corr = df[num_cols].corr()

plt.figure(figsize=(14,10))
sns.heatmap(corr, cmap="coolwarm", center=0, annot=False)
plt.title("Feature Correlation Matrix")
plt.show()

# --- 7. Correlation with target ---
target_corr = corr["target_encoded"].drop("target_encoded").sort_values(key=abs, ascending=False)

print("\nTop correlations with koi_disposition:")
print(target_corr)

plt.figure(figsize=(8,6))
target_corr.plot(kind="barh")
plt.title("Correlation of Features with koi_disposition")
plt.xlabel("Correlation")
plt.show()
```

```python
import pandas as pd
import os
import re

# -----------------------------
# 1. Load catalog
# -----------------------------
df = pd.read_csv("kepler_core.csv")

print("Original dataset shape:", df.shape)

# -----------------------------
# 2. Get available KIC IDs from lightcurve folder
# -----------------------------
lc_folder = "output_lightcurves"  # change to your folder path
files = os.listdir(lc_folder)

# Extract KIC IDs from filenames like lightcurve_0_KIC6020753.csv
kic_ids = []
for f in files:
    match = re.search(r"KIC(\d+)", f)
    if match:
        kic_ids.append(int(match.group(1)))

kic_ids = set(kic_ids)
print(f"Found {len(kic_ids)} KIC IDs with lightcurves")

# -----------------------------
# 3. Filter dataset to keep only those rows
# -----------------------------
df_matched = df[df["kepid"].isin(kic_ids)].copy()
print("Filtered dataset shape:", df_matched.shape)

# -----------------------------
# 4. Encode target
# -----------------------------
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_matched["target_encoded"] = le.fit_transform(df_matched["koi_disposition"])

label_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
print("\nLabel mapping:", label_mapping)

# -----------------------------
# 5. Save filtered dataset
# -----------------------------
df_matched.to_csv("kepler_with_lightcurves.csv", index=False)
print("\n✅ Saved kepler_with_lightcurves.csv (only rows with lightcurves available)")
```

```python
import pandas as pd
import os, re

df = pd.read_csv("kepler_core.csv")
print("Unique kepid in catalog:", df["kepid"].nunique())

# Extract KIC IDs from lightcurve filenames
lc_folder = "output_lightcurves/"  # change path
files = os.listdir(lc_folder)
kic_ids = set(int(re.search(r"KIC(\d+)", f).group(1)) for f in files if "KIC" in f)

print("Unique KIC IDs in folder:", len(kic_ids))

# Compare overlap
catalog_ids = set(df["kepid"].unique())
overlap = catalog_ids & kic_ids
missing_in_catalog = kic_ids - catalog_ids
missing_in_folder = catalog_ids - kic_ids

print("Overlap IDs:", len(overlap))
print("KICs missing in catalog:", len(missing_in_catalog))
print("KICs missing in folder:", len(missing_in_folder))

# # Save for inspection
# pd.Series(sorted(list(missing_in_catalog))).to_csv("missing_in_catalog.csv", index=False)
# pd.Series(sorted(list(missing_in_folder))).to_csv("missing_in_folder.csv", index=False)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Load dataset
df = pd.read_csv("kepler.csv")

print("🔍 Original shape:", df.shape)
print("🎯 Target distribution:\n", df["koi_disposition"].value_counts())

# -----------------------
# 1. Handle missing values
# -----------------------
# Drop columns with >50% missing
threshold = 0.5 * len(df)
df = df.dropna(thresh=threshold, axis=1)

# Fill remaining missing numerical values with median
for col in df.columns:
    if df[col].dtype in ["float64", "int64"]:
        df[col] = df[col].fillna(df[col].median())

print("✅ Missing values handled")

# -----------------------
# 2. Encode target
# -----------------------
# Binary: CONFIRMED = 1, others = 0
df["target"] = df["koi_disposition"].apply(lambda x: 1 if x == "CONFIRMED" else 0)

print("🎯 Target encoding done")
print(df["target"].value_counts())

# -----------------------
# 3. Drop unused columns
# -----------------------
drop_cols = ["rowid", "kepid", "kepoi_name", "koi_disposition"]
df_clean = df.drop(columns=drop_cols)

# -----------------------
# 4. Save train-ready CSV
# -----------------------
df_clean.to_csv("train_ready.csv", index=False)
print("💾 Saved cleaned dataset to train_ready.csv")
print("📊 Final shape:", df_clean.shape)
```
